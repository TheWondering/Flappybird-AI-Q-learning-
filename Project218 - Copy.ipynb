```python
import os, sys
import gymnasium as gym
import random
import numpy as np
from tqdm import tqdm
import text_flappy_bird_gym
```

```python
env = gym.make('TextFlappyBird-v0', height = 15, width = 20, pipe_gap = 4)
obs = env.reset()
print(obs)

def initialize_q_table(obs_space1, obs_space2, action_space):
    Qtable = np.zeros((obs_space1, obs_space2, action_space))
    return Qtable

def greedy_policy(Qtable, state):
    action = np.argmax(Qtable[state][:])
    return action

def epsilon_greedy_policy(Qtable, state, epsilon):
  random_num = random.uniform(0,1)
  if random_num > epsilon:
    action = greedy_policy(Qtable, state)
  else:
    action = env.action_space.sample()
  return action
    
def train(n_training_episodes, min_epsilon, max_epsilon, decay_rate, env, Qtable):
    for episode in tqdm(range(n_training_episodes)):
        epsilon = min_epsilon + (max_epsilon - min_epsilon)*np.exp(-decay_rate*episode)
        state = env.reset()[0]
        step = 0
        done = False
        
        while True:
            action = epsilon_greedy_policy(Qtable, state, epsilon)
            new_state, reward, done, _, info = env.step(action)
            Qtable[state[0]][state[1]][action] = Qtable[state[0]][state[1]][action] + learning_rate * (reward + gamma * np.max(Qtable[new_state[0]][new_state[1]]) - Qtable[state[0]][state[1]][action])
            if done:
                break
            state = new_state
            
    return Qtable
```
